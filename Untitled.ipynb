### Apply capping using Feature-engine a Python library with multiple transformers to engineer and select features to use in machine learning models. 

Gaussian limits:

right tail: mean + 2* std

left tail: mean - 2* std

```python
from feature_engine.outliers import Winsorizer
```

```python
windsoriser = Winsorizer(capping_method='gaussian', # choose iqr for IQR rule boundaries or gaussian for mean and std
                          tail='both', # cap left, right or both tails 
                          fold=2,
                          variables=['Variance','Skewness','Kurtosis','Entropy'])

windsoriser.fit(df)
```

```python
df_w = windsoriser.transform(df)
```

```python
df_w.describe()
```

```python
%matplotlib inline
df_w.hist(bins=50, figsize=(20,20))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import GradientBoostingClassifier

from xgboost.sklearn import XGBClassifier
from xgboost import plot_importance
from sklearn.metrics import plot_precision_recall_curve
from yellowbrick.classifier import (ROCAUC,
    ClassificationReport,
    DiscriminationThreshold,
    ConfusionMatrix,
    PrecisionRecallCurve,
)
from yellowbrick.model_selection import FeatureImportances, feature_importances
from sklearn.metrics import (precision_score, roc_auc_score, recall_score, confusion_matrix,f1_score,auc,average_precision_score,
                             roc_curve, precision_recall_curve, accuracy_score,explained_variance_score, plot_precision_recall_curve)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   df_w.drop(labels=['Class'], axis=1), # drop the target
    df_w['Class'], # just the target
    test_size=0.2,
    random_state=0)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model_lr = LogisticRegression(C=1e5, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=400, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False).fit(X_train, y_train)

pred_train_lr = model_lr.predict(X_train)
pred_test_lr = model_lr.predict(X_test)


accuracy_train_lr = accuracy_score(pred_train_lr,y_train)
accuracy_test_lr = accuracy_score(pred_test_lr,y_test)


fpr_train, tpr_train, _ = roc_curve(np.array(y_train), model_lr.predict_proba(X_train)[:,1])
auc_train = auc(fpr_train,tpr_train)

fpr_test, tpr_test, _ = roc_curve(np.array(y_test), model_lr.predict_proba(X_test)[:,1])
auc_test = auc(fpr_test,tpr_test)

print(accuracy_train,accuracy_test,auc_train,auc_test)
```

```python
disp = plot_precision_recall_curve(model_lr, X_test, y_test)
y_score=model_lr.predict(X_test)
average_precision = average_precision_score(y_test, y_score)
disp.ax_.set_xbound(0.0,1)
disp.ax_.set_ybound(0.5,1)
disp.ax_.set_title('2-class Precision-Recall curve: '
                   'AP={0:0.2f}'.format(average_precision))
```

```python
tree = DecisionTreeClassifier(random_state=42)

tree.fit(X_train, y_train)

pred_train_tr = tree.predict(X_train)
pred_test_tr = tree.predict(X_test)


accuracy_train_tr = accuracy_score(pred_train_tr,y_train)
accuracy_test_tr = accuracy_score(pred_test_tr,y_test)


fpr_tr_train, tpr_tr_train, _ = roc_curve(np.array(y_train), tree.predict_proba(X_train)[:,1])
auc_train = auc(fpr_tr_train,tpr_tr_train)

fpr_tr_test, tpr_tr_test, _ = roc_curve(np.array(y_test),tree.predict_proba(X_test)[:,1])
auc_test = auc(fpr_tr_test,tpr_tr_test)

print(accuracy_train_tr,accuracy_test_tr,auc_train,auc_test)
```

```python
disp = plot_precision_recall_curve(tree, X_test, y_test)
y_score=tree.predict(X_test)
average_precision = average_precision_score(y_test, y_score)
disp.ax_.set_xbound(0.0,1)
disp.ax_.set_ybound(0.5,1)
disp.ax_.set_title('2-class Precision-Recall curve: '
                   'AP={0:0.2f}'.format(average_precision))
```
